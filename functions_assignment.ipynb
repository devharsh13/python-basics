Question 1. What is the difference between a function and a method in Python?

Answers.  A function is a block of code which can be reuse that performs a specific task. It is defined using the def keyword and can be called independently, without being associated with an object.

A method is a function that is associated with an object and is called using the object. It is defined inside a class and operates on the data within that class.

```python
# function example

def add(a, b):
    return a * b

print(add(5, 10))  # Output: 15
```

```python
# Method Example

class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

obj = Calculator(5, 10)
print(obj.add())  # Output: 15
```

Question 2. Explain the concept of function arguments and parameters in Python ?

Answers. Function Arguments and Parameters in Python

Parameters are variables defined in a function that that are placed between the parenthesis.

Arguments are actual values passed to a function during a function call.

Types of Function Arguments:

Positional Arguments – Values are passed in the same order as parameters.

Default Arguments – Assigns a default value if no value is provided.

Keyword Arguments – Arguments are passed using parameter names.

Arbitrary Arguments (*args) – Allows passing multiple values as a tuple.

Arbitrary Keyword Arguments (**kwargs) – Allows passing multiple key-value pairs as a dictionary.

```python
# 1. Positional Arguments
def greet(name, age):
    print(f"Hello, {name}, Age: {age}")
greet("Alice", 25)

# 2. Default Arguments
def greet(name, age=18):
    print(f"Hello, {name}, Age: {age}")
greet("Bob")

# 3. Keyword Arguments
greet(age=30, name="Charlie")

# 4. Arbitrary Arguments (*args)
def add(*nums):
    print(sum(nums))
add(1, 2, 3)

# 5. Arbitrary Keyword Arguments (**kwargs)
def info(**data):
    print(data)
info(name="David", age=40)
```

Question 3. What are the different ways to define and call a function in Python ?

Answers.  Different Ways to Define and Call a Function in Python
Standard Function Definition and Call
A function is defined using the def keyword and called by its name.
It does not accept any parameters and is used to execute a block of code.

Function with Parameters
Functions can accept parameters that are passed when the function is called. These parameters can be used inside the function.

Function with Default Parameters
Default values can be assigned to parameters. If no value is passed when the function is called, the default value is used.

Lambda Function
A lambda function is a short, anonymous function defined using the lambda keyword. It's useful for short operations.

Recursive Function
A recursive function is one that calls itself in order to solve a problem, typically with a base case to stop the recursion.

```python
# 1. Standard Function Definition and Call
def greet():
    print("Hello!")

greet()  # Function call

# 2. Function with Parameters
def add(a, b):
    return a + b

result = add(5, 10)  # Function call
print(result)

# 3. Function with Default Parameters
def greet(name="Guest"):
    print(f"Hello, {name}!")

greet()  # Uses default value
greet("Alice")  # Function call with argument

# 4. Lambda (Anonymous) Function
square = lambda x: x * x
print(square(4))  # Function call

# 5. Recursive Function
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  # Function call
```

Question 5. What are iterators in Python and how do they differ from iterables ?

Answers.  An iterator is an object in Python that allows you to iterate over a sequence of elements (like lists, tuples, or strings). It implements two key methods:

__iter__() – Returns the iterator object itself.
__next__() – Returns the next item in the sequence and raises StopIteration when the sequence ends.
An iterable, on the other hand, is any Python object capable of returning its members one at a time, allowing it to be looped over in a for-loop. Examples include lists, tuples, strings, etc. An iterable must implement the __iter__() method to return an iterator.





```python
# Example : Creating an iterable (List)
my_list = [1, 2, 3]

# Iterating over the iterable
for item in my_list:
    print(item)  # Output: 1 2 3

# Example 2: Creating an iterator
iter_obj = iter(my_list)  # Convert the iterable into an iterator

# Iterating over the iterator using __next__()
print(next(iter_obj))  # Output: 1
print(next(iter_obj))  # Output: 2
print(next(iter_obj))  # Output: 3
# print(next(iter_obj))  # Raises StopIteration error after the last element
```

Question 6. Explain the concept of generators in Python and how they are defined ?

Answers.  A generator in Python is a special type of iterator that generates values one at a time as they are needed, instead of storing all values in memory . This makes generators more memory efficient, especially when dealing with large datasets.

Generators are defined using the yield keyword inside a function. When the function is called, it returns a generator object without executing the function immediately. Each time next() is called on the generator, the function runs until it hits the yield statement, returning the value, and then pauses at that point.

```python
# Example : Defining a generator function
def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # Yield value and pause the function
        count += 1

# Using the generator
counter = count_up_to(3)  # Create a generator object

# Get values one at a time
print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
# print(next(counter))  # Raises StopIteration error when no more items
```

Question 7. What are the advantages of using generators over regular functions?

Answers.  Generators provide several advantages over regular functions, especially when working with large datasets or when memory efficiency is crucial.

> Memory Efficiency

> Suitable for Infinite Sequences

> State Retention

```python
# Regular function that returns all values at once
def get_squares(n):
    result = []
    for i in range(1, n+1):
        result.append(i * i)
    return result

squares = get_squares(5)
print(squares)  # Output: [1, 4, 9, 16, 25]

# Generator function that yields one value at a time
def generate_squares(n):
    for i in range(1, n+1):
        yield i * i  # Yield one value at a time

squares_gen = generate_squares(5)
for square in squares_gen:
    print(square)  # Output: 1 4 9 16 25
```

Question 10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?

Answer. map : apply or define a function of each time in a collection. it is generally use in list and tuples .

filter : Selects elements from an data type (e.g., list, tuple) that meet a specific condition or we can say that to perform a specific operation in a specific elements .

reduce : the reduce() function is a function that helps combine all elements of a collection like a list or tuple into a single value by applying a specified function repeatedly.

syntax of reduce:

from functool import reduce

reduce(function, iterable[, initializer])






```python
from functools import reduce

# Original list of integers
numbers = [3, -1, 2, -7, 5, -3, 4]

# Step 1: Filter out negative numbers
positive_numbers = filter(lambda x: x >= 0, numbers)

# Step 2: Square each of the remaining numbers
squared_numbers = map(lambda x: x ** 2, positive_numbers)

# Step 3: Sum all the squared numbers
sum_of_squares = reduce(lambda x, y: x + y, squared_numbers)

print(sum_of_squares)  # Output: 54
```

Question 4. Write a Python function that checks if a given number is prime or not from 1 to 200 ?

```python
def prime(n):
      if n <= 1:
        return False
      for i in range(2, n):
        if n % i == 0:
            return False
      return True

number = int(input("Enter a number between 1 and 200: "))
if prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

Question 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms ?

```python
class FibonacciIterator:
    def __init__(self, n):

        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):

        if self.count >= self.n:
            raise StopIteration

        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib

n_terms = 10
fib_iter = FibonacciIterator(n_terms)
for num in fib_iter:
    print(num, end=" ")
```

Question 6. Write a generator function in Python that yields the powers of 2 up to a given exponent ?

```python
def powers_of_2(exponent):

    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
max_exponent = 5
for power in powers_of_2(max_exponent):
    print(power, end=" ")
```

 Question 7. Implement a generator function that reads a file line by line and yields each line as a string ?


```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

file_path = "example.txt"
for line in read_file_line_by_line(file_path):
    print(line)
```

Question 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit ?

```python
celsius_to_fahrenheit = lambda c: (c * 9/5) + 32

celsius_temperatures = [0, 20, 37, 100, -10]

fahrenheit_temperature = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(f"Temperatures in Fahrenheit: {fahrenheit_temperature}")
```

Question 10. Create a Python program that uses `filter()` to remove all the vowels from a given string ?
